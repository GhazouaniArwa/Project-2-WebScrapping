```python
#import pacakage 
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
#write url for working 
url = "https://feeds.bbci.co.uk/news/rss.xml"
```

```python
#import code page 
response = requests.get(url)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
items=soup.findAll('item')
```

```python
title=soup.findAll('title')
```

```python
#show items 
items
```

```python
title
```

```python
item= items[3]
```

```python
item.title.text
```

```python
item.description.text
```

```python
#extraire information 
new_items=[]
```

```python
for i in items:
    new_i={}
    new_i['title']=i.title.text
    new_i['description']=i.description.text
    new_i['pubdate']=i.pubdate.text 
    new_items.append(new_i)
```

```python
new_items
```

```python
#convert list on data
df=pd.DataFrame(new_items,columns=['title','description','pubdate'])
```

```python
df
```

```python
#Data preparation
titles = df['title']
date = df['pubdate']
```

```python
# Text cleaning
import re
proper_titles = [re.sub(r'[^a-zA-Z0-9\s]', '', title) for title in titles]
```

```python
proper_titles
```

```python
#Date conversion
# Conversion des dates en objets de date/heure
date  = pd.to_datetime(date)
```
